# KPI Views – Pharmacy Count by Disease

This notebook creates reporting views based on the Gold aggregate table:

`capstone.gold.pharmacy_count_disease_gold`

These views support business reporting related to disease-level pharmacy coverage and disease distribution analysis.

The views are designed for:
- dashboard development
- pharmacy coverage comparison across diseases
- top/bottom disease availability analysis

```python
%sql
CREATE OR REPLACE VIEW capstone.kpi.kpi_pharmacy_count_by_disease AS
SELECT
    disease_id,
    disease_name,
    region,
    pharmacy_count
FROM capstone.gold.pharmacy_count_disease_gold;
```

```python
%sql
CREATE OR REPLACE VIEW capstone.kpi.kpi_top_diseases_by_pharmacy_coverage AS
SELECT
    disease_id,
    disease_name,
    pharmacy_count
FROM capstone.gold.pharmacy_count_disease_gold
ORDER BY pharmacy_count DESC;
```

```python
%sql
CREATE OR REPLACE VIEW capstone.kpi.kpi_disease_pharmacy_coverage_band AS
SELECT
    disease_id,
    disease_name,
    pharmacy_count,
    CASE
        WHEN pharmacy_count >= 300 THEN 'High Coverage'
        WHEN pharmacy_count >= 150 THEN 'Medium Coverage'
        ELSE 'Low Coverage'
    END AS coverage_band
FROM capstone.gold.pharmacy_count_disease_gold;
```

```python
%sql
CREATE OR REPLACE VIEW capstone.kpi.kpi_inventory_freshness AS
SELECT
    state,
    region,
    MAX(last_updated) AS latest_update,
    MIN(last_updated) AS oldest_update,
    COUNT(*) AS total_records,
    COUNT(DISTINCT pharmacy_id) AS pharmacy_count
FROM capstone.gold.disease_medication_pharmacy_gold
GROUP BY
    state,
    region;
```

```python
%sql
CREATE OR REPLACE VIEW capstone.kpi.kpi_availability_by_region AS
SELECT
    region,
    SUM(total_quantity) AS total_quantity_available,
    SUM(pharmacy_count) AS pharmacy_count,
    COUNT(DISTINCT medication_id) AS medication_count
FROM capstone.gold.medication_availability_region_gold
GROUP BY region;
```

```python
%sql
CREATE OR REPLACE VIEW kpi_medication_price_comparison_by_state as 
SELECT
  state,
  medication_name,
  avg_price,
  min_price,
  max_price
FROM
  capstone.gold.medication_price_state_gold
ORDER BY
  state,
  medication_name
```

```python
%sql
CREATE OR REPLACE VIEW capstone.kpi.kpi_inventory_freshness AS

SELECT
    state,
    region,

    COUNT(*) AS total_inventory_records,

    MAX(last_updated) AS latest_inventory_update,

    MIN(last_updated) AS oldest_inventory_update,

    DATEDIFF(current_timestamp(), MAX(last_updated)) AS days_since_last_update

FROM capstone.gold.disease_medication_pharmacy_gold

GROUP BY
    state,
    region;
```
