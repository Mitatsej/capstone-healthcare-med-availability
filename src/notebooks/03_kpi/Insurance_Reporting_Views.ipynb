# Gold Insurance Reporting Views

This notebook creates reporting views from the insurance-enriched Gold table:

`capstone.gold.disease_medication_pharmacy_insurance_gold`

These views support extended business analysis related to:
- insurance coverage by disease
- average copay by state
- prior authorization requirements
- specialty drug distribution

The purpose of these views is to simplify dashboard development and provide additional analytical insights on insurance-related medication access.

```python
%sql
CREATE OR REPLACE VIEW capstone.kpi.kpi_insurance_coverage_by_disease AS
SELECT
    disease_id,
    disease_name,
    coverage_status,
    COUNT(*) AS total_records,
    COUNT(DISTINCT insurance_id) AS insurance_plan_count,
    COUNT(DISTINCT medication_id) AS medication_count
FROM capstone.gold.disease_medication_pharmacy_insurance_gold
WHERE insurance_id IS NOT NULL
GROUP BY
    disease_id,
    disease_name,
    coverage_status;
```

```python
%sql
CREATE OR REPLACE VIEW capstone.kpi.kpi_average_copay_by_state AS
SELECT
    state,
    ROUND(AVG(copay), 2) AS avg_copay,
    ROUND(AVG(coinsurance), 2) AS avg_coinsurance,
    COUNT(DISTINCT insurance_id) AS insurance_plan_count,
    COUNT(DISTINCT medication_id) AS medication_count
FROM capstone.gold.disease_medication_pharmacy_insurance_gold
WHERE insurance_id IS NOT NULL
  AND copay IS NOT NULL
GROUP BY state;
```

```python
%sql
CREATE OR REPLACE VIEW capstone.kpi.kpi_specialty_drug_distribution_by_region AS
SELECT
    region,
    is_specialty_drug,
    COUNT(*) AS total_records,
    COUNT(DISTINCT medication_id) AS medication_count,
    COUNT(DISTINCT insurance_id) AS insurance_plan_count
FROM capstone.gold.disease_medication_pharmacy_insurance_gold
WHERE insurance_id IS NOT NULL
GROUP BY
    region,
    is_specialty_drug;
```
